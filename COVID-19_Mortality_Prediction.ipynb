```python
#NEURAL NETWORK
```

```python
X = df.drop('outcome',axis = 1).values
y = df['outcome'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 25)
```

```python
#MODEL1- 19,10,1 with drop out 0.5
```

```python
model = Sequential()

model.add(Dense(19,activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(10,activation = 'relu'))
model.add(Dropout(0.5))
#Binary Classification
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer = 'adam')
```

```python
model.fit(x = X_train,y= y_train,epochs = 500,validation_data = (X_test,y_test),callbacks = [early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
NNpredictions = model.predict_classes(X_test)
print(classification_report(y_test,NNpredictions))
```

```python
accuracy ['Neural Network']=accuracy_score(y_test,NNpredictions)
# model.evaluate(X_test, y_test, verbose=1)

#print the accuracy
accuracy
```
